```python
data_cleaned = data[~data.isna().any(axis=1)]
data_cleaned = data_cleaned[~data_cleaned.isin(['*']).any(axis=1)]
data_cleaned = data_cleaned[~data_cleaned.isin(['Unknown']).any(axis=1)]
```

```python
data_cleaned
```

```python
to_bin_columns= ['TOTAL NUMBER OF HIV DIAGNOSES' 'HIV DIAGNOSES PER 100,000 POPULATION', 'TOTAL NUMBER OF AIDS DIAGNOSES', 'AIDS DIAGNOSES PER 100,000 POPULATION']
for to_bin in to_bin_columns:
    data_cleaned[to_bin]=pd.qcut(np.array(data_cleaned[to_bin]).astype('float32'), 4,  duplicates='drop',labels=["Low", "Medium", "High"])
```

```python
data_cleaned.head()
```

```python
data_cleaned = data_cleaned.rename(columns={"HIV DIAGNOSES PER 100,000 POPULATION": "HIV PER 100K", "AIDS DIAGNOSES PER 100,000 POPULATION": "AIDS PER 100K"})
```

```python
data_cleaned.shape
```

```python
for column in data_cleaned.columns:
    data_cleaned[column] = data_cleaned[column].apply(lambda x: column+"="+ str(x))
```

```python
data_cleaned.columns
```

```python
data_cleaned = data_cleaned.drop(columns=['YEAR','TOTAL NUMBER OF CONCURRENT HIV/AIDS DIAGNOSES', 'PROPORTION OF CONCURRENT HIV/AIDS DIAGNOSES AMONG ALL HIV DIAGNOSES'])
```

```python
data_cleaned.to_csv("INTEGRATED-DATASET.csv", index=False)
```
