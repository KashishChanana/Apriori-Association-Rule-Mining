```python
data_cleaned = data[~data.isna().any(axis=1)]
data_cleaned = data_cleaned[~data_cleaned.isin(['*']).any(axis=1)]
data_cleaned = data_cleaned[~data_cleaned.isin(['Unknown']).any(axis=1)]
```

```python
to_bin_columns= ['TOTAL NUMBER OF HIV DIAGNOSES', 'HIV DIAGNOSES PER 100,000 POPULATION', 'TOTAL NUMBER OF CONCURRENT HIV/AIDS DIAGNOSES', 'PROPORTION OF CONCURRENT HIV/AIDS DIAGNOSES AMONG ALL HIV DIAGNOSES', 'TOTAL NUMBER OF AIDS DIAGNOSES', 'AIDS DIAGNOSES PER 100,000 POPULATION']
for to_bin in to_bin_columns:
    data_cleaned[to_bin]=pd.cut(np.array(data_cleaned[to_bin]).astype('float32'), 3, labels= ['Low', 'Medium', 'High'],ordered=True)
```

```python
for column in data.columns:
    data_cleaned[column] = data_cleaned[column].apply(lambda x: column+"="+ str(x))
```
